```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
 class Data:
        
        #Year
        YEAR =["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"]
        
        #RACE TYPE(distance, dart)
        RACE_TYPE = ' ダ1200m (右)\n'
        
        #SAPPORO, HAKODATE, FUKUSHIMA, NIGATA, TOKYO, NAKAYAMA, CHYUKYO, KYOTO, HANSHIN, KOKURA
        PLACE = ["01","02","03","04","05","06","07","08","09","10"]
        
        #Times
        KAI = ["01", "02", "03", "04", "05","6"]
        
        #Day
        DAY = ["01", "02", "03", "04", "05", "06", "07", "08", "09"]
        
        #Race_No
        NO = ["01","02","03","04","05","06","07","08","09","10", "11", "12"]
        
        def Get_url(self,race):
            
            #Get url
            url = 'https://race.netkeiba.com/?pid=race&id=c'+race+'&mode=result'

            #Responseオブジェクト生成
            response = requests.get(url)
    
            #文字化け防止
            response.encoding = response.apparent_encoding 
    
            #BeautifulSoupオブジェクト生成
            soup = BeautifulSoup(response.text, "html.parser") 
 
            try:
                #' ダ1200m (右)\n'確認   
                state=soup.find('div', class_='RaceData01').text
           
                if(state.split("/")[1]==self.RACE_TYPE):
                    #テーブル取得
                    table = soup.find('tbody')
                    rows = table.findAll("tr")
                    print(url)
                    with open("data/"+race+".csv", "w", encoding='utf-8') as file:
                        writer = csv.writer(file)
                        for row in rows:
                            csvRow = []
                            for cell in row.findAll(['td', 'th']):
                                csvRow.append(cell.get_text())
                            writer.writerow(csvRow) 
            except AttributeError:
                print('Error')
```

```python
data=Data()
for year in data.YEAR:
    for kai in data.KAI:
        for day in data.DAY:
            for no in data.NO:
                race=year+data.PLACE[5]+kai+day+no
                data.Get_url(race)
```
